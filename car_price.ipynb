## **CAR PRICE PREDICTION**

```python
import pandas as pd

car_data = pd.read_csv(r'https://raw.githubusercontent.com/amankharwal/Website-data/master/CarPrice.csv')
car_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(car_data.corr())
```

```python
features = ['wheelbase', 'carlength','carwidth', 'enginesize', 'boreratio','horsepower','price']
new_data = car_data[features]
new_data.head()
```

```python
new_data.describe()
```

```python
sns.pairplot(new_data,hue='price', vars=['wheelbase', 'carlength','carwidth', 'enginesize', 'boreratio','horsepower'])
```

```python
new_features = ['wheelbase', 'carlength','carwidth', 'enginesize', 'boreratio','horsepower']
X = car_data[new_features]
y = car_data['price']

from sklearn.model_selection import train_test_split 
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor

car_model = DecisionTreeRegressor()
car_model.fit(train_X, train_y)

predication = car_model.predict(test_X) 
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print('Mean absolute error: ', round(mean_absolute_error(test_y, predication),3))
print('Mean squared error: ', round(mean_squared_error(test_y, predication),3))
print('r2 score: ', round(r2_score(test_y, predication),3))
```
